# Data Analysis and Visualization

### <b> What is the most popular year for WeRateDogs? </b>

The above chart shows the number of Tweets per year for the WeRateDog's Trend on Twitter. The most popular year for this trend was 2016. In 2016 there were 741 Tweets for WeRateDogs. In 2015 there were 432 Tweets and in 2017 there were 283 Tweets. From this analysis it is clear that the WeRateDogs Trend reached the height of its popularity in 2016.

### <b> What is the breakdown of Dog Rankings? </b>

On WeRateDogs there is a ranking system among the dogs posted. Dogs can fall under 4 different rankings: pupper, doggo, puppo, and floofer. The distribution of these ranks can be viewed in the pie chart above. The most popular rank for a dog is pupper with nearly 67% of the ranks being this. Puppers are usually very cute young dogs! There isn't much suprise that the vast majority of photos being posted are of cute puppies, everyone loves them. There is no need to worry though the older dogs get shown love too. The next most popular ranking is doggo which is usually a full grown dog.

### <b> What Platform provided the most Tweets? </b>

Twitter is provided on multiple platforms and users are able to send Tweets from multiple devices. Tweets sent during the WeRateDogs trend were predominantly sent from Twitter for iPhone. Twitter for iPhone accounted for 98% of the Tweets.

### <b> Is there a correlation between favorites and retweets? </b>

Is there correlation between favorites and retweets? To explore this it was hypothesized that popular tweets receive a large amount of favorites and retweets. This was explored using the Pearson's Correlation Coefficient which measures the linear correlation between two variables. A value of 1 would show a total positive linear correlation. The coefficient between favorites and retweets is 0.928 which shows a very strong correlation between these two variable proving the hypothesis true. This is visualized in the plot above.
